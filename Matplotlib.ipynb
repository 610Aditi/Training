```python
font_dic={'color':'white','fontfamily':'monospace','fontsize':20,'fontweight':'extra bold'}
```

```python
fig = plt.figure(facecolor='red')
ax = fig.add_axes([0, 0, 1, 1], facecolor='black')
x = np.linspace(-5, 5, 10)

ax.plot(x, x**2, 'r--o', label='square')
ax.plot(x, x**3, 'w:s', label='cube')
ax.legend()
ax.set_title("Anatomy of figure", fontsize=30, fontdict=font_dic)
ax.set_xlabel("x-axis", fontsize=20, fontdict=font_dic)
ax.set_ylabel("y-axis", fontsize=20, fontdict=font_dic)
ax.set_xticks(np.arange(-6,6))
ax.set_xticklabels([chr(65+i) for i in range(12)], rotation=45,fontdict=font_dic)
ax.set_yticks(np.arange(-130, 130, 20))
#xmin, xman, ymin, ymax
ax.grid()
plt.show()
```

# It is also possible to set a log scale for one or both axes . this functioality is in fact only one application of a more general transformation system in Matplotlib. Each of the axs scales are set seperately using set+xscale and set_ysacle metheds which accept one parameter (with the value of 'log' in this case )do we have interfaces in python

```python
fig, axes =plt.subplots(1,2,figsize = (10,4))

axes[0].plot(x,x**2,x,np.exp(x))
axes[0].set_title("normal scale ")

axes[1].plot(x,x**2,x,np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logrithmic scale(y)")
```

```python
fig ,ax = plt.subplots(figsize =(10,5))
ax.plot(x,x**2,x,x**3,lw=2)
ax.set_xticks([-5,-2,0,10,15])
ax.set_xticklabels([r'\alpha$',r'$\beta$',r'\$gamma$',r'\$delta$',r'$epsilon$'],fontsize=18)
yticks = [0,50,100,150]
ax.set_yticks([25,100,125,200,225])
ax.set_yticks(yticks)
ax.set_yticklabels(['$%.1f$' % y for y in yticks],fontsize=18)   # use LATEX formatted labels
```
